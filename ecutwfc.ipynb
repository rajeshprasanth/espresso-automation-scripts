```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline
plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
```

```python
x, a = np.loadtxt('Li3Sb.psl.pbe.nvdw.dat', delimiter=' ', unpack=True)
x, b = np.loadtxt('Li3Sb.psl.pbe.vdw.dat', delimiter=' ', unpack=True)
x, c = np.loadtxt('Li3Sb.psl.pz.nvdw.dat', delimiter=' ', unpack=True)
x, d = np.loadtxt('Li3Sb.psl.pz.vdw.dat', delimiter=' ', unpack=True)

plt.plot(x, a, marker = '<', markersize=5, label='PSL/PBE/No vdw-DF')
plt.plot(x, b, marker = '>', markersize=5, label='PSL/PBE/vdw-DF')
plt.plot(x, c, marker = '^', markersize=5, label='PSL/PZ/No vdw-DF')
plt.plot(x, d, marker = 'v', markersize=5, label='PSL/PZ/vdw-DF')

plt.xlabel('ecutwfc (Ry)')
plt.ylabel('Etot (Ry)')
plt.title("Li3Sb ecutwfc Convergence Test")
plt.legend(frameon=False)
plt.show()
```
